```python
import pandas as pd
df=pd.read_csv("tamil_nadu_budget.csv")
df["Growth_Percentage"]=df["Growth_Percentage"].fillna(0)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(df["Sector"],df["Allocation_Crores"],color='royalblue')
plt.title("Budget allocation for Tamil Nadu(2025-26)")
plt.xlabel("Government sectors")
plt.ylabel("Allocation in Crores")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
explode_settings = (0.1, 0, 0, 0, 0, 0, 0)

plt.pie(df["Allocation_Crores"], labels=df["Sector"],autopct="%1.1f%%", startangle=140, explode=explode_settings,colors=plt.cm.Set2.colors)
plt.title("Tamil Nadu Budget Distribution by Percentage (2025-26)", fontweight='bold')

plt.show()
```
